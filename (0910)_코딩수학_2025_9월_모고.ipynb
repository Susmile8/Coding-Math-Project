Process 1. 함수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-2,13,1000)
k=5
y1=np.abs(x-k)
y2=x**2-11*x+18

plt.figure(figsize=(10,6))
plt.plot(x,y1,label=r'$y=|x-k|$', color='blue')
plt.plot(x,y2,label=r'$y=x^2-11x+18$', color='green')
plt.title(f'function1: $y=|x-k|$ & function2: $y=x^2-11x+18$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

Process 2. 풀이과정 및 답 구하기

```python
answer=[]
for k in range(0,13):
  y1_min=k-4
  y1_max=k+4
  y2_min=2
  y2_max=9
  inter_min=max(y1_min,y2_min)
  inter_max=min(y1_max,y2_max)
  count=0
  for x in range(int(inter_min)+1, int(inter_max)):
    count+=1
  if count==3:
    answer.append(k)
sum_k=sum(answer)
print(answer)
print(sum_k)
```

Process 1. 함수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.linspace(0.01,1,300)
x2=np.linspace(1.01,3,200)
f1=[-np.log(x)/np.log(3) for x in x1]
f2=[np.log(x)/np.log(3) for x in x2]

plt.plot(x1,f1,label=r'$f(x)=\log_\frac{1}{3} x$', color='blue')
plt.plot(x2,f2,label=r'$f(x)=\log_3 x$',color='green')
plt.axvline(1,color='gray',linestyle='-')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('logarithmic functions graph')
plt.grid(True)
plt.legend()
plt.show()
```

Process 2. 풀이과정 및 답 구하기

```python
from sympy import symbols,Eq,log,solve,Rational

x=symbols('x',real=True)

# Case1: 0 < x <= 1/3
# f(x) = -log3(x), f(3x) = -log3(3x)
# 방정식: -log3(x) - log3(3x) = 3 → log3(3x^2) = -3
eq1 = Eq(log(3 * x**2, 3), -3)
sol1 = [s for s in solve(eq1, x) if 0 < s <= Rational(1, 3)]

# Case2: 1/3 < x <= 1
# f(x) = -log3(x), f(3x) = log3(3x)
# f(x) + f(3x) = -log3(x) + log3(3x) = log3(3x / x) = log3(3)
# log3(3) = 1 < 3 → 조건 불만족 → 실근 없음
sol2 = []

# Case3: x > 1
# f(x) = log3(x), f(3x) = log3(3x)
# log3(x) + log3(3x) = log3(3x^2) = 3
eq3 = Eq(log(3 * x**2, 3), 3)
sol3 = [s for s in solve(eq3, x) if s > 1]

print("Case1: 0 < x <= 1/3:",sol1)
print("Case2: 1/3 < x <= 1:",sol2)
print("Case3: x > 1:",sol3)
print("Answer:", sol1[0]+sol3[0])
sol1[0]+sol3[0]
```
